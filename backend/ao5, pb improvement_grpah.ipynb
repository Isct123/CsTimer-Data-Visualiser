```python
import pb_checker
import utils.preprocess_solves as pf
```

```python
session = pf.load_all_sessions("data/suku.txt")[0]
```

```python
session
```

```python
averages = {}
for i in range(len(session.solves)-4):
    solves = session.solves[i:i+5]
    when = solves[-1].date
    ao5 = [i.time for i in solves]
    ao5 = sorted(ao5)
    averages.update({when: sum(ao5[1:-1]) / 3})
    
averages
```

```python
session.solves[-1]
```

```python
import matplotlib.pyplot as plt

plt.plot(averages)
plt.title("Line Plot of Large List")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
from datetime import datetime
ao5_dict = averages

# Step 1: Sort and parse the dictionary
timestamps = sorted(ao5_dict.keys())
dates = [datetime.strptime(ts, "%Y-%m-%d %H:%M:%S") for ts in timestamps]
ao5_values = [ao5_dict[ts] for ts in timestamps]

# Step 2: Plot
plt.figure(figsize=(12, 6))
plt.plot(dates, ao5_values, linewidth=1.5)
plt.title("ao5 Over Time (Improvement)")
plt.xlabel("Date")
plt.ylabel("ao5 Time (s)")
plt.grid(True)
plt.tight_layout()
plt.gcf().autofmt_xdate()  # Auto-format x-axis for dates
plt.show()
```
